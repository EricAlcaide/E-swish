```python
import json
import numpy as np 
import matplotlib.pyplot as plt
```

```python
filenames = ["log.txt", "log (1).txt", "log (2).txt"]
plots = []

for log in filenames:
    results = []
    # Extract raw logs
    with open(log) as f:
        content = f.readlines()
    #Clean them up
    for i, line in enumerate(content):
        s = line[12:]
        json_acceptable_string = s.replace("'", "\"")
        d = json.loads(json_acceptable_string)
        results.append(d["test_acc"])
        
    plots.append(results)
    print("\n", results)
```

```python
print(max(plots[0]), max(plots[1]), max(plots[2]))
```

```python
%matplotlib notebook
plt.figure()
for result in plots:
    plt.plot(100-np.array(result))

plt.ylim(0, 45)
plt.grid()
plt.show()
```
